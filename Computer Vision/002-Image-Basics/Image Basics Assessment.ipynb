```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
img = cv2.imread('../DATA/dog_backpack.jpg')
```

```python
plt.imshow(img)
```

```python
fix_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_image)
```

```python
new_img = cv2.flip(fix_image, 0)
plt.imshow(new_img)
```

```python
plt.imshow(fix_image)
```

```python
img_rectangle = cv2.rectangle(fix_image, pt1=(200, 400), pt2=(600, 700), color=(255, 0, 0), thickness=10)
plt.imshow(img_rectangle)
```

```python
dog_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(dog_image)
```

```python
cv2.rectangle(dog_image, pt1=(200, 400), pt2=(600, 700), color=(255, 0, 0), thickness=10)
plt.imshow(dog_image)
```

```python
img_triangle = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
vertices = np.array([[450, 400], [250, 700], [600, 700]], dtype=np.int32)
```

```python
pts = vertices.reshape((-1, 1, 2))
cv2.polylines(img_triangle, [pts], isClosed=True, color=(255, 0, 0), thickness=10)
plt.imshow(img_triangle)
```

```python
img_triangle_filled = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
vertices = np.array([[425, 400], [250, 700], [600, 700]], dtype=np.int32)
```

```python
pts = vertices.reshape((-1, 1, 2))
cv2.fillPoly(img_triangle_filled, [pts], color=(0, 0, 255))
plt.imshow(img_triangle_filled)
```
