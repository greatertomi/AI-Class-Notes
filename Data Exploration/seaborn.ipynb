```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Distplot: helps to see the distribution of your data along 1 column

sns.distplot(tips['total_bill'], kde=False, bins=30)
```

### Joinplot

Jointplot allows you to match up two distplot for bivariate data

_Bivariate data is a data that has two variables_

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

### Pairplot

Pairplot plots pair-wise relationship across an entire dataframe (atleast for the numerical columns). It also supports a hue argument for categorical columns.

```python
sns.pairplot(tips, hue='sex')
```

### Rug Plot

Rugplot draws a dash line for everypoint on the univariate data given to it.

```python
sns.rugplot(tips['total_bill'])
```

## Categorical Plots

### Barplot

Barplot allows you to aggregate the categorical data based of some functions and by default that's the mean. You can think of it as a visualization of the groupby action.

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)

# The estimator parameter is optional. You can pass in whatever function you want include the ones you create
```

### Countplot

Countplot is the same as barplot except that the estimator is exclusively counting the number of occurrences.

```python
sns.countplot(x='sex', data=tips)
```

### Boxplot

It shows the distribution of quantitative data in a way that hopefully facilitates comparison between the variables.

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

### Violinplot

It works like boxplot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

### StripPlot

Stripplot is a scatterplot based of the category

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

### Swarmplot

It combines a stripplot and a violinplot. It functions very similar to stripplot, just that the points are adjusted so that they don't overlap.

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

### Factorplot

Is the more general plot of all the plot above. You can recreate any plot by specifying it in the kind argument.

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

## Matrix Plots

Matrix plots only works well when your data is in a matrix form. That is that the row and column have headings that match. To do that you could use
- data.corr()
- data.pivot_table(*args)

_For *args, pass in the neccesary arguments_

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot=True)
```

```python
flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')

sns.heatmap(fp)
```

### Clustermap

Uses hierarchical clustering to create a cluster version of the heatmap

```python
sns.clustermap(fp)
```
