```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional method of generating a plot
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1) # Parameters are num_of_rows, num_of_cols, plot_number
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object oriented method

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Parameters are left, bottom, width and height

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('Plot 1')

axes[1].plot(y, x)
axes[1].set_title('Plot 2')

plt.tight_layout() # Handles overlap
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

### Save a figure

```python
fig.savefig('my_picture.png', dpi=250)
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend()
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='#FF8C00', linewidth=3, alpha=0.8, linestyle='-.', marker='+', markersize=16)

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
