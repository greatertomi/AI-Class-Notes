```python
import numpy as np
x = np.arange(0, 100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# print(x)
# print(y)
# print(z)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, .4, .4])

ax1.plot(x, y, z)
ax2.plot(x, y, z)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y, lw=3, ls='--', color='blue')
axes[1].plot(x, z, lw=4, color='red')
```
